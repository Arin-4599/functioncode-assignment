6. Explain the concept of generators in Python and how they are defined?
   - generator function uses yield and normal function uses return
   - In Python, generators are a special type of iterable that allow you to iterate over a sequence of values, but they generate the values on the fly (lazily) as they are needed, rather than computing all of them at once and storing them in memory.
   - The first time the generator is iterated, the function starts executing until it encounters yield. After that, it suspends execution and returns the yielded value.

   - Each subsequent call to next() on the generator resumes execution from the point where it was paused.
   - eg:- def square_number_generators(n):
    for i in range(n):
        yield i**2
    
print(square_number_generators(10)) # yield is creating a generator similar to iterator object

gen = square_number_generators(10)
print(next(gen))
print(next(gen))
print(next(gen))
